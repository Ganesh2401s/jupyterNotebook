```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dataframe = {
    "Rooms" : [2, 2, 2, 1, 2, 3, 1, 1, 1, 3, 3, 2, 2], 
    "Area" : [1100, 800, 900, 250, 1000, 1200, 400, 250, 375, 900, 1286, 600, 800],
    "Bathroom" : [2, 1, 2, 1, 2, 2, 1, 1, 1, 2, 2, 1, 1],
    "Furnishing_Status" : ["Unfurnished", "Semi-Furnished", "Furnished", "Unfurnished", "Semi-Furnished", "Semi-Furnished", "Unfurnished", "Furnished", "Unfurnished", "Unfurnished", "Furnished", "Semi-Furnished", "Unfurnished" ], 
    "Rent" : [10000, 16000, 22000, 5000, 23000, 25000, 7000, 6500, 6000, 8500, 35000, 8000, 12000]
}

House_Rent = pd.DataFrame(dataframe)
House_Rent
```

```python
max_rent = House_Rent["Rent"].max()
House_Rent[House_Rent["Rent"] == max_rent]
```

```python
total_area = House_Rent["Area"].sum()
total_rent = House_Rent["Rent"].sum()
print(f"Total area is {total_area} and Total rent is {total_rent}.")
```

```python
one_room = House_Rent["Rooms"][House_Rent["Rooms"] == 1].count()
two_room = House_Rent["Rooms"][House_Rent["Rooms"] == 2].count()
three_room = House_Rent["Rooms"][House_Rent["Rooms"] == 3].count()

print(f"The houses with 1, 2, 3 rooms are {one_room}, {two_room}, {three_room} in number respectively.")
```
