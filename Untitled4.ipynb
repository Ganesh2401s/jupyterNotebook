### First made a figure object using figure instance method of matplotlib.pyplot submodule, then created a AxesSubplot object using figure object using add_subplot instance method which takes nrows, ncols, index as parameters, created Reactangle, circle patches using matplotlib.pyplot submodule then added the patch in axessubplot object using **add_patches** instance method

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

fig = plt.figure(facecolor="gray", edgecolor="b")
ax = fig.add_subplot(1, 1, 1)
rect1 = plt.Rectangle((0.2, 0.4), 0.4, 0.2, color="b", alpha=0.3 )
rect2 = plt.Rectangle((0.2, 0.6), 0.4, 0.2, color="g", alpha=0.3 )
circle = plt.Circle((0.8, 0.6), 0.2, color="r", alpha=0.2)
ax.add_patch(rect1)
ax.add_patch(rect2)
ax.add_patch(circle)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#figure object
fig = plt.figure()
#AxesSubplot Object 
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)
# plt.plot(np.random.randn(50).cumsum(), "k--")

ax1.plot(np.random.randn(50).cumsum(), "k-", drawstyle="steps-post", label="steps-post")
# ax1.legend(loc="best")
ax1.plot(np.random.randn(50).cumsum(), "r--", label="default")
ax1.legend(loc="best", fontsize="small")
ax2.hist(np.random.randn(50), bins=30, color="g")
ax3.scatter(np.random.randn(50), np.arange(0, 500, 10) * 2.5)
ax4.bar(np.random.randn(50), np.arange(0, 500, 10) * 2.5)
```
