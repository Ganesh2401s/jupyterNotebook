```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

fig = plt.figure(figsize=(7,5))  #figure object
ax1 = fig.add_subplot(2, 2, 1)  #AxesSubplot Objects
ax2 = fig.add_subplot(2, 2, 2)  #AxesSubplot Objects
ax3 = fig.add_subplot(2, 2, 3)  #AxesSubplot Objects
ax4 = fig.add_subplot(2, 2, 4)  #AxesSubplot Objects

rect = plt.Rectangle((0.2, 0.2), 0.4, 0.6, color="b", alpha=0.5)
circ = plt.Circle((0.5, 0.5), 0.2, color="m", alpha=0.3)
poly = plt.Polygon([[0.2, 0.3], [0.4, 0.5], [0.2, 0.6], [0.3, 0.7]], color="g", alpha=0.3)

ax1.add_patch(rect)
ax2.add_patch(circ)
ax3.add_patch(poly)
 

plt.plot(np.random.rand(100).cumsum(), "b--")
plt.show()
```
