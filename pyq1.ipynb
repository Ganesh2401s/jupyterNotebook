```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


data = {
    "Name" : ["Ram", "Ravi", "Reena", "Rita", "Rishi", "Romi"], 
    "Age" : [15, 23, 32, 20, 33, 21],
    "Weight" : [45.6, 34.9, 45.6, 60.7, 54.7, 34.6], 
    "Height" : [140, 160, 145, 155, 170, 144]
}

df = pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(8, 10))
plt.subplot(2, 2, 1)
plt.scatter(df["Age"], df["Height"], marker="*")
plt.title("Scatter Plot between Age and Height")
plt.xlabel("Age")
plt.ylabel("Height")


plt.subplot(2, 2, 2)
plt.barh(df["Name"], df["Weight"], color="red", alpha=0.5)
plt.title("Horizontal bar plot")
plt.xlabel("Weight")
plt.ylabel("Name")

plt.subplot(2, 2, 3)
plt.scatter(df["Age"], df["Height"], marker="*")
plt.title("Scatter Plot between Age and Height")
plt.xlabel("Age")
plt.ylabel("Height")

plt.subplot(2, 2, 4)
plt.scatter(df["Age"], df["Height"], marker="*")
plt.title("Scatter Plot between Age and Height")
plt.xlabel("Age")
plt.ylabel("Height")

plt.suptitle("Data Analysis")
plt.savefig('analysis.png', dpi=100, bbox_inches='tight', transparent=True)
plt.show()
```

#### Write a python code to create a figure and a subplot using matplotlib functions. Plot a rectangle of size 3.5 x 8.5 at a point (2.0, 7.0), a circle of radius 2.5 at point (7.0, 2.0) as patches in the subplot, functions for plottig. Set the color of rectangle as "Green" and color of circle as "Blue", Set the x-scale and y-scale to 1-10. Import appropriate libraries.

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create a figure and a subplot
fig, ax = plt.subplots()

# Create a rectangle at (2.0, 7.0) with width 3.5 and height 8.5
rectangle = patches.Rectangle((2.0, 7.0), 3.5, 8.5, color='green')

# Create a circle centered at (7.0, 2.0) with radius 2.5
circle = patches.Circle((7.0, 2.0), 2.5, color='blue')

# Add the rectangle and circle to the subplot
ax.add_patch(rectangle)
ax.add_patch(circle)

# Set the x and y axis limits
ax.set_xlim(1, 10)
ax.set_ylim(1, 10)

# Set aspect ratio to equal for accurate shapes
ax.set_aspect('equal')

# Display the plot
plt.title("Rectangle and Circle Patches")
# plt.grid(True)
plt.show()
```
