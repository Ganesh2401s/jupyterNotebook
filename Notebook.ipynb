```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml

#Titanic dataset
titanic = fetch_openml('titanic', version=1, as_frame=True)
df = titanic.data

df.sample(5) # printing 5 random rows from titanic datset 
```

```python
df['survived'] = titanic.target.astype(int)
df['fare'] = pd.to_numeric(df['fare'], errors='coerce')
df['age'] = pd.to_numeric(df['age'], errors='coerce')
df.dropna(subset=['sex', 'age', 'fare', 'survived'], inplace=True)
```

```python
# a. Plot bar chart to show the frequency of survivors and non-survivors for male and female passengers
plt.figure(figsize=(8, 6))
sns.countplot(x='sex', hue='survived', data=df, palette='coolwarm')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.title('Survivors and Non-Survivors by Gender')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
# b. Draw a scatter plot for age vs fare
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='fare', hue='survived', data=df, alpha=0.7, palette='coolwarm')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
# c. Compare density distribution for age and passenger fare
plt.figure(figsize=(8, 6))
sns.kdeplot(df['age'], label='Age', fill=True, color='blue')
sns.kdeplot(df['fare'], label='Fare', fill=True, color='red')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Density Distribution for Age and Fare')
plt.legend()
plt.show()
```

```python
# d. Use a pair plot to show pairwise bivariate distribution
sns.pairplot(df[['age', 'fare', 'survived']], hue='survived', palette='coolwarm')
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml

# Load Titanic dataset
titanic = fetch_openml('titanic', version=1, as_frame=True)
df = titanic.data

# Ensure proper datatypes
df['survived'] = titanic.target.astype(int)
df['fare'] = pd.to_numeric(df['fare'], errors='coerce')
df['age'] = pd.to_numeric(df['age'], errors='coerce')
df['pclass'] = pd.to_numeric(df['pclass'], errors='coerce')
df.dropna(subset=['age', 'fare', 'survived', 'pclass'], inplace=True)
```

```python
# a. Find total number of passengers with age less than 30
total_under_30 = df[df['age'] < 30].shape[0]
print(f'Total passengers with age less than 30: {total_under_30}')
```

```python
# b. Find total fare paid by passengers of first class
total_fare_first_class = df[df['pclass'] == 1]['fare'].sum()
print(f'Total fare paid by first-class passengers: {total_fare_first_class}')
```

```python
# c. Compare number of survivors of each passenger class
plt.figure(figsize=(8, 6))
sns.countplot(x='pclass', hue='survived', data=df, palette='coolwarm')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.title('Survivors by Passenger Class')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```
