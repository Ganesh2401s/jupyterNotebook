```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

x = np.array([5, 100, 150])
y = np.array([20, 150, 62])
font_title = {"family": "arial", "color": "blue", "size": 15 }
#Line graph: 
# plt.plot(x-axis, y-axis, marker='o', ms=10, color="")
# plt.plot(x, y, marker="D", color ="red")

# 'D:r' = Diamond + dotted line + Red
# 'o-.b' = Circled marker + Dashed-dotted line + Blue accent color
# 'H-g' : Hexagon marker + Solid line + green Color
# 'd--m' : small diamond marker + Dashed line + Magenta Color
plt.plot(x, y, 'o-.m', ms=5, mfc="green", mec="black")
plt.title("Title of the line graph", fontdict = font_title, loc="left")
plt.xlabel("X-axis label" )
plt.ylabel("Y-axis label")
plt.grid(axis='y', color="blue", alpha=0.1, linestyle='--', linewidth=1)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.polynomial.polynomial import Polynomial

# Original data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

# Fit a linear model (degree 1)
p = Polynomial.fit(x, y, 1)

# New x-values for extrapolation (beyond the original range)
x_extrapolated = np.linspace(0, 7, 100)
y_extrapolated = p(x_extrapolated)

# Plot
plt.plot(x, y, 'o', label="Original Data")
plt.plot(x_extrapolated, y_extrapolated, label="Extrapolated Line", linestyle='--')
plt.title("Extrapolation with Matplotlib")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Line parameters
intercept = 0.3
slope = 0.24  # from the calculation above

# Generate x values (including negative range to show x-axis intersection)
x = np.linspace(-2, 2, 100)
y = slope * x + intercept

# Plotting
plt.figure(figsize=(8, 5))
plt.plot(x, y, label=f"y = {slope:.2f}x + {intercept}", color='blue')
plt.axhline(0, color='black', linewidth=0.5)  # x-axis
plt.axvline(0, color='black', linewidth=0.5)  # y-axis
plt.title("Line Graph with Intercept 0.3 and x-axis crossing at -1.25")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```
