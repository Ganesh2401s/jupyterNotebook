```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
df = sns.load_dataset('Iris')
print(df)
```

```python
# categorical and numerical columns
categorical_features = df.select_dtypes(include=['object']).columns
numeric_features = df.select_dtypes(include=['number']).columns

print(f"Categorical features: {len(categorical_features)} -> {list(categorical_features)}")
print(f"Numeric features: {len(numeric_features)} -> {list(numeric_features)}")
```

```python
# correlation matrix
correlation_matrix = df[numeric_features].corr()
print(correlation_matrix)

# Visualize correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

df_reduced = df.drop(columns=['petal_width'])
print(df_reduced.head())
```

```python
# Five-number summary for each attribute
summary = df_reduced.describe().loc[['min', '25%', '50%', '75%', 'max']]
print(summary)
```

```python
# Boxplots for each numeric attribute
for col in numeric_features:
    if col != 'petal_width':  # Skip removed feature
        sns.boxplot(x=df[col])
        plt.title(f'Boxplot of {col}')
        plt.show()
```
